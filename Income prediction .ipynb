```python
# Splitting the data into X and Y and use of Ransac Regressor we are going to handle the Outliers 
X = train_data.drop(('revised_net_salary'),axis=1)
Y = train_data['revised_net_salary']
```

```python
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor()
ransac.fit(X,Y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
x = X[inlier_mask]
y = Y[inlier_mask]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
test_data.marital_status = label.fit_transform(test_data.marital_status)
```

```python
test_data.dtypes
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_valid ,Y_train,Y_valid = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
X_test = test_data
```

```python
from sklearn.preprocessing import StandardScaler
Std = StandardScaler()
```

```python
X_train = Std.fit_transform(X_train)
X_valid = Std.transform(X_valid)
```

```python
X_test = Std.transform(X_test)
```

```python
def model(X_train,Y_train):
    #Linear regression 
    from sklearn.linear_model import LinearRegression
    Linear = LinearRegression()
    Linear.fit(X_train,Y_train)
    
    #XG boost Regressor 
    from xgboost import XGBRegressor
    xg = XGBRegressor()
    xg.fit(X_train,Y_train)
    
    # printing the scores 
    print('[0] Linear Regression',Linear.score(X_train,Y_train))
    print('[1] XG Boost Regression',xg.score(X_train,Y_train))
    
    
    return Linear , xg 
```

```python
models = model(X_train,Y_train)
```

```python
Y_prediction = models[1].predict(X_valid)
```

```python
from sklearn.metrics import r2_score
print(r2_score(Y_valid,Y_prediction))
```

```python
Y_prediction_test = models[1].predict(X_test)
```
